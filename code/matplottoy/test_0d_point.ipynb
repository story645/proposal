```python
%matplotlib inline
import numpy as np

import matplotlib.pyplot as plt

import data
import channels
import artists
```

```python
table = data.Section()
```

```python
table.view().keys()
```

```python
cmap = {'true':'green', 'false':'red'}
transforms = {'y': ('v1', channels.Position()),       
              'x': ('v2', channels.Position()),
            'facecolors': ('v3', channels.NominalColor(cmap))} 
```

```python
fig, ax = plt.subplots()

artist = artists.Point(table, transforms)
ax.add_artist(artist)
ax.set(xlim = (table.FB.F[transforms['x'][0]].min, 
               table.FB.F[transforms['x'][0]].max), 
       ylim = (table.FB.F[transforms['y'][0]].min, 
               table.FB.F[transforms['y'][0]].max),       
       xlabel = transforms['x'][0], 
       ylabel = transforms['y'][0], 
       aspect='equal')
```

```python
### fails on facecolors
fig, ax = plt.subplots()

artist = artists.Line(table, transforms)
ax.add_artist(artist)
ax.set(xlim = (table.FB.F[transforms['x'][0]].min, 
               table.FB.F[transforms['x'][0]].max), 
       ylim = (table.FB.F[transforms['y'][0]].min, 
               table.FB.F[transforms['y'][0]].max),       
       xlabel = transforms['x'][0], 
       ylabel = transforms['y'][0], 
       aspect='equal')
```
