```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


import matplottoy.datasources.array as da
import matplottoy.artists.point as ap
```

```python
%reload_ext autoreload
%autoreload 2
```

```python
# regularized array
arr = np.cumsum(np.tile([1.,2.,.03, 5.], reps=100).reshape(100,4), axis=0)
```

```python
ds = da.ArrayPoint(arr)
ds.info
```

```python
fig, ax = plt.subplots()
ds = da.ArrayPoint(arr)
artist = ap.Point(ds)
ax.add_artist(artist)
ax.set(xlim = (ds.info['x']['min'], ds.info['x']['max']),
       ylim = (ds.info['y']['min'], ds.info['y']['max']))
```

```python
fig, ax = plt.subplots()
ds = da.ArrayPoint(arr, encodings={'x':1, 'y':3})
artist = ap.Point(ds)
ax.add_artist(artist)
ax.set(xlim = (ds.info['x']['min'], ds.info['x']['max']),
       ylim = (ds.info['y']['min'], ds.info['y']['max']))
```

```python
fig, ax = plt.subplots()
ds = da.ArrayPoint(arr, encodings={'x':3, 'y':1, 's':2})
artist = ap.Point(ds)
ax.add_artist(artist)
ax.set(xlim = (ds.info['x']['min'], ds.info['x']['max']),
       ylim = (ds.info['y']['min'], ds.info['y']['max']))
```
