```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from blume.table import table
```

```python
# regularized array
df = pd.DataFrame(np.cumsum(np.tile([1.,2.,.03,5.], reps=100).reshape(100,4), axis=0), 
                  columns = ['odds', 'evens','threes', 'fives'])
dfs = df[:5]
```

```python
fig, axd = plt.subplot_mosaic([['table', 'image'],
                               ['table', 'scatter']], constrained_layout=True)
table(axd['table'], cellText=dfs.astype('str').values, colLabels=dfs.columns, bbox=(0, 0, 1, 1))
axd['image'].matshow(dfs.values[:4])
axd['scatter'].scatter(dfs['odds'], dfs['evens'], c=dfs['threes'], s=dfs['fives'])
fig.savefig("viz_same.png")
```
